```python
n = int(input("enter the number"))

i = 2
while i <= n:
    print(i)
    i += 2
```

```python
#Method 5 (Using List Comprehension)
n = int(input("enter the number"))

even = [i for i in range(1, n+1) if i % 2 == 0]

print(even)
```

```python
def even(n):
    for i in range(2, n+1, 2):
        print(i)

n = int(input("Enter N: "))
even(n)
```

```python
def even(n):
    even_list = []
    for i in range(2, n+1, 2):
        even_list.append(i)
    return even_list

n = int(input("enter the number"))
print(even(n))
```

```python
def even(n):
    if n == 0:
        return
    even(n-1)
    if n % 2 == 0:
        print(n)

n = int(input("enter the number"))
even(n)
```
