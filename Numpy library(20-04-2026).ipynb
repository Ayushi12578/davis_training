```python
def get_even_factors(num):
    factors = []
    for i in range(1, int(num**0.5) + 1):
        if num % i == 0:
            if i % 2 == 0:
                factors.append(i)
            if (num // i) % 2 == 0 and (num // i) != i:
                factors.append(num // i)
    factors.sort()
    return factors

given_number = 30
list_even_factors = get_even_factors(given_number)
array = np.array(list_even_factors)

print(f"Given number: {given_number}, Even factors: {list_even_factors}")
print(array)
```
